```python
a = footprint_accounts.groupby(['country'], as_index=False).agg({'carbon': 'sum'}).sort_values(by=['carbon'], ascending=False)
b = a.reset_index(drop=True)
b
```

```python
top_carbon_emitting_countries = list(b['country'][1:20])
top_carbon_emitting_values    = list(b['carbon'][1:20])
```

```python
# libraries
import matplotlib.pyplot as plt
import squarify    # pip install squarify (algorithm for treemap)
 
# If you have 2 lists
f, ax = plt.subplots(figsize=(20, 15))
squarify.plot(sizes=top_carbon_emitting_values, label = top_carbon_emitting_countries, alpha=.7 )

plt.axis('off')
plt.show()
 
# If you have a data frame?
# import pandas as pd

# squarify.plot(sizes=b['carbon'], label=b['country'], alpha=.8 )
# plt.axis('off')
# plt.show()
```
